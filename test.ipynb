```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/Hotel Reservations.csv')

print("Primeras 5 filas del DataFrame:")
print(df.head())

print("\nInformación del DataFrame:")
df.info()

print("\nEstadísticas descriptivas:")
print(df.describe())

print("\nValores nulos por columna:")
print(df.isnull().sum())

# Ejemplo de visualización: Histograma de 'lead_time'
plt.figure(figsize=(10, 6))
sns.histplot(df['lead_time'], bins=50, kde=True)
plt.title('Distribución de Lead Time')
plt.xlabel('Lead Time')
plt.ylabel('Frecuencia')
plt.show()

# Ejemplo de visualización: Conteo de 'booking_status'
plt.figure(figsize=(8, 5))
sns.countplot(x='booking_status', data=df)
plt.title('Estado de las Reservas')
plt.xlabel('Estado de Reserva')
plt.ylabel('Conteo')
plt.show()
```
